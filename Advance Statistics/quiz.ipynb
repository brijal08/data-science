```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import zscore
from factor_analyzer import FactorAnalyzer

import warnings
warnings.filterwarnings("ignore")
```

```python
pca_data = pd.read_csv('FB.csv')
pca_data.head()
```

```python
pca_data.info()
```

```python
pca_data.describe().T
```

```python
pca_data_new = pca_data.drop(['status_id'],axis= True)
df_scalled = pca_data_new.apply(zscore)
df_scalled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=6,random_state=123)
pca_score = pca.fit_transform(df_scalled)
pca_score
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
lstNumeric = list(pca_data.select_dtypes(exclude='object').columns)
df_pca_loading = pd.DataFrame(pca.components_,columns=list(lstNumeric))
df_pca_loading
```

```python
pca = PCA(n_components=12,random_state=123)
pca_score = pca.fit_transform(df_scalled)
pca_score[1]
```
